```python
def compute_reciprocals(values):
    output=np.empty(len(values))
    for i in range(len(values)):
        output[i]=1.0/values[i]
    return output
```

# Boolean operators

So we see that there are 29 days with rainfall between 0.5 and 1.0 inches.

```python
# Another approach
np.sum(~((inches<=0.5)|(inches>=1)))
```

The following table summarizes the bitwise Boolean operators and their equivalent ufuncs:


```python
print("Number of days without rain:   ", np.sum(inches==0))
print("Number of days with rain:      ", np.sum(inches!=0))
print("Days with more than 0.5 inches:", np.sum(inches>0.5))
print("Rainy days with <0.2 inches    ", np.sum((inches>0)&(inches<0.2)))
```

# Using the Keywords and/or Versus the Operators &/|

When you use and or or, it’s equivalent to asking Python to treat the object as a single
Boolean entity. In Python, all nonzero integers will evaluate as True. Thus:

```python
bool(42), bool(0)
```

```python
bool(42 and 0)
```

```python
bool(42 or 0)
```

When you use & and | on integers, the expression operates on the bits of the element,
applying the and or the or to the individual bits making up the number:

```python
bin(42)
```

```python
bin(59)
```

```python
bin(42 & 59)
```

```python
bin(42|59)
```

When you have an array of Boolean values in NumPy, 
this can be thought of as a string of bits where 1 = True and 0 = False

```python
A = np.array([1,0,1,0,1,0], dtype=bool)
B = np.array([1,1,1,0,1,1], dtype=bool)
A|B
```

Using or on these arrays will try to evaluate the truth or falsehood of the entire array object, 
which is not a well-defined value.

```python
#A or B
```

---------------------------------------------------------------------------
ValueError                                Traceback (most recent call last)
<ipython-input-277-ea2c97d9d9ee> in <module>
----> 1 A or B

ValueError: The truth value of an array with more than one element is ambiguous. Use a.any() or a.all()

```python
x=np.arange(10)
```

```python
(x>4)&(x<8)
```

Trying to evaluate the truth or falsehood of the entire array will give the same ValueError.

In short, and and or perform a single Boolean evaluation on an entire object, 
while & and | perform multiple Boolean evaluations on the content.

modify complicated subsets of an array’s values.

```python
[x[3], x[7], x[2]]
```

```python
ind=[3,7,4]
x[ind]
```

With fancy indexing, the shape of the result reflects the shape of the index arrays
rather than the shape of the array being indexed.

```python
ind=np.array([[3,7],[4,5]])
x[ind]
```

# Example: Binning Data

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set() # for plot style

np.random.seed(42)
x=np.random.rand(100)
x
```

```python
# compute a histogram by hand
bins=np.linspace(-5,5,20)
counts=np.zeros_like(bins)
```

```python
bins
```

```python
counts
```

```python
# find the appropriate bin for each x
i=np.searchsorted(bins,x)
i
```

```python
#add 1 to each of these bins
np.add.at(counts,i,1)
```

```python
counts
```

```python
# plot the results
plt.plot(bins, counts, linestyle='steps');
```

Of course, it would be silly to have to do this each time you want to plot a histogram.
This is why Matplotlib provides the plt.hist() routine, which does the same in a single line:

```python
plt.hist(x, bins, histtype='step');
```

```python
print("NumPy routine:")
%timeit counts, edges = np.histogram(x, bins)
```

```python
print("Custom routine:")
%timeit np.add.at(counts, np.searchsorted(bins, x), 1)
```

# Sorting Arrays

```python
# numpy.argmin(array, axis = None, out = None) : Returns indices of the min element of the array in a particular axis.
# https://www.geeksforgeeks.org/numpy-argmin-python/

import numpy as np

def selection_sort(x):
    for i in range(len(x)):
        swap = i + np.argmin(x[i:])
        (x[i], x[swap]) = (x[swap], x[i])
    return x
```

```python
x = np.array([2, 1, 4, 3, 5])
selection_sort(x)
```

# Structured Data: NumPy’s Structured Arrays

```python
import numpy as np
```

```python
name = ['Alice', 'Bob', 'Cathy', 'Doug']
age = [25, 45, 37, 19]
weight = [55.0, 85.5, 68.0, 61.5]
```

```python
x = np.zeros(4, dtype=int)
```

```python
# Use a compound data type for structured arrays
data = np.zeros(4, dtype={'names':('name','age','weight'),
                          'formats':('U10','i4','f8')})
```

```python
print(data.dtype)
```

```python
data['name']=name
data['age']=age
data['weight']=weight
print(data)
```

```python
# Get all names
data['name']
```

```python
# Get first row of data
data[0]
```

```python
# Get the name from the last row
data[-1]['name']
```

```python
# Using Boolean masking
# Get names where age is under 30
data[data['age']<30]['name']
```
