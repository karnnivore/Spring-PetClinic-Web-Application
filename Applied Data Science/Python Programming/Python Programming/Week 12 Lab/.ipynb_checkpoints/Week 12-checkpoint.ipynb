```python
#Key, Value
```

```python
fruits = {
    'red':'apple',
    'blue':'berries',
    'orange':'tangerine'
}
```

```python
for key in fruits:
    print("Key is "+key)
```

```python
for x in fruits:
    print("Key is "+x)
```

```python
for key in fruits.keys():
    print("Key is "+key)
```

```python
for value in fruits.values():
    print("Value is "+value)
```

```python
for key, value in fruits.items():
    print("Key is {:s}, Value is {:s}".format(key, value))
```

```python
rates ={
    'morgage':{
        'home':5.5,
        'auto':3.5,
    },
    'loan':{
        'secured':7.5,
        'unsecured':8.5,
    },
}
```

```python
len(rates)
```

```python
print(rates['morgage']['home'])
```

```python
print(rates['loan']['unsecured'])
```

```python
rates['morgage']['auto']=6.2
```

```python
print(rates['morgage']['auto'])
```

```python
for i in rates:
    for j in rates[i]:
        print(i, ":", j, "=", rates[i][j])
```
